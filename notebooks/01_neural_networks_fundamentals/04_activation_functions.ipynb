# Activation Functions

Activation functions introduce non-linearity to neural networks, allowing them to learn complex patterns.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define common activation functions

def sigmoid(x):
    return 1 / (1 + np.exp(-np.clip(x, -500, 500)))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

def softmax(x):
    e_x = np.exp(x - np.max(x, axis=0, keepdims=True))
    return e_x / np.sum(e_x, axis=0, keepdims=True)

# Plot activation functions
x = np.linspace(-5, 5, 200)

fig, axes = plt.subplots(2, 3, figsize=(15, 8))

# Sigmoid
axes[0, 0].plot(x, sigmoid(x), 'b-', linewidth=2)
axes[0, 0].set_title('Sigmoid', fontweight='bold')
axes[0, 0].grid(True, alpha=0.3)
axes[0, 0].set_ylim([-0.1, 1.1])

# Tanh
axes[0, 1].plot(x, tanh(x), 'g-', linewidth=2)
axes[0, 1].set_title('Tanh', fontweight='bold')
axes[0, 1].grid(True, alpha=0.3)
axes[0, 1].set_ylim([-1.1, 1.1])

# ReLU
axes[0, 2].plot(x, relu(x), 'r-', linewidth=2)
axes[0, 2].set_title('ReLU', fontweight='bold')
axes[0, 2].grid(True, alpha=0.3)

# Leaky ReLU
axes[1, 0].plot(x, leaky_relu(x), 'purple', linewidth=2)
axes[1, 0].set_title('Leaky ReLU', fontweight='bold')
axes[1, 0].grid(True, alpha=0.3)

# Derivatives
h = 1e-5
sigmoid_deriv = (sigmoid(x + h) - sigmoid(x - h)) / (2 * h)
axes[1, 1].plot(x, sigmoid_deriv, 'b-', linewidth=2, label='Sigmoid')
axes[1, 1].plot(x, sigmoid_deriv * (tanh(x)**2 - 1), 'g-', linewidth=2, label='Tanh')
axes[1, 1].set_title('Activation Derivatives', fontweight='bold')
axes[1, 1].legend()
axes[1, 1].grid(True, alpha=0.3)
axes[1, 1].set_ylim([0, 0.3])

axes[1, 2].axis('off')

plt.tight_layout()
plt.show()
```

## Activation Function Properties

| Function | Formula | Derivative | Range | Use Cases |
|----------|---------|-----------|-------|----------|
| Sigmoid | $\sigma(x) = \frac{1}{1+e^{-x}}$ | $\sigma(x)(1-\sigma(x))$ | (0, 1) | Binary classification output |
| Tanh | $\tanh(x)$ | $1 - \tanh^2(x)$ | (-1, 1) | Hidden layers (faster convergence than sigmoid) |
| ReLU | $\max(0, x)$ | 1 if x>0, else 0 | [0, ∞) | Most common for hidden layers |
| Leaky ReLU | $\max(\alpha x, x)$ | $\alpha$ if x<0, else 1 | (-∞, ∞) | Prevents dying ReLU problem |
| Softmax | $\frac{e^{x_i}}{\sum e^{x_j}}$ | $p_i(1-p_i)$ | (0, 1) | Multi-class classification output |

```python
# Example: Softmax for multi-class classification
logits = np.array([[2.0], [1.0], [0.1]])  # Raw output scores
probabilities = softmax(logits)

print("Logits:", logits.flatten())
print("Softmax probabilities:", probabilities.flatten())
print("Sum of probabilities:", np.sum(probabilities))
```
