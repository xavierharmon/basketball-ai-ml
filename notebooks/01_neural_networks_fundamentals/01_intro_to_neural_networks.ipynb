# Introduction to Neural Networks

## What is a Neural Network?

A neural network is a computational model inspired by biological neurons in the brain. It consists of interconnected nodes (neurons) organized in layers that learn to make predictions from data.

### Key Concepts:
- **Neuron**: Basic computational unit that receives inputs, applies weights, adds bias, and passes through activation
- **Layer**: Collection of neurons processing data together
- **Weights**: Learnable parameters that scale inputs
- **Bias**: Learnable parameters that shift activation
- **Activation Function**: Non-linear transformation that enables learning complex patterns

```python
import numpy as np
import matplotlib.pyplot as plt

# Visualize a simple neuron
fig, ax = plt.subplots(figsize=(10, 6))

# Draw a simple feedforward network
# Input layer
ax.scatter([0, 0, 0], [0.25, 0.5, 0.75], s=500, c='blue', label='Input Layer')

# Hidden layer
ax.scatter([1, 1], [0.375, 0.625], s=500, c='green', label='Hidden Layer')

# Output layer
ax.scatter([2], [0.5], s=500, c='red', label='Output Layer')

# Draw connections
for i in [0.25, 0.5, 0.75]:
    for j in [0.375, 0.625]:
        ax.plot([0, 1], [i, j], 'k-', alpha=0.2)

for j in [0.375, 0.625]:
    ax.plot([1, 2], [j, 0.5], 'k-', alpha=0.2)

ax.set_xlim(-0.3, 2.3)
ax.set_ylim(0, 1)
ax.set_aspect('equal')
ax.legend(loc='upper right')
ax.set_title('Simple Feedforward Neural Network', fontsize=14, fontweight='bold')
ax.axis('off')
plt.tight_layout()
plt.show()
```

## How Neural Networks Work

### 1. Forward Pass
Data flows from input → hidden layers → output

**Mathematical formula for one neuron:**
$$z = w_1 x_1 + w_2 x_2 + ... + w_n x_n + b$$
$$a = \sigma(z)$$

Where:
- $x_i$ = input features
- $w_i$ = weights
- $b$ = bias
- $\sigma()$ = activation function
- $a$ = output

```python
# Simple neuron computation
def simple_neuron(x, weights, bias, activation_fn=None):
    """Compute output of a single neuron"""
    z = np.dot(weights, x) + bias  # Linear combination
    
    if activation_fn is None:
        return z
    else:
        return activation_fn(z)  # Apply activation

# Example: single neuron with 2 inputs
inputs = np.array([0.5, 0.3])
weights = np.array([0.2, 0.8])
bias = 0.1

output = simple_neuron(inputs, weights, bias)
print(f"Input: {inputs}")
print(f"Weights: {weights}")
print(f"Bias: {bias}")
print(f"Linear output (z): {output:.4f}")
```

### 2. Learning Process
Neural networks learn by:
1. Making predictions (forward pass)
2. Computing error (loss)
3. Computing gradients (backpropagation)
4. Updating weights (optimization)
