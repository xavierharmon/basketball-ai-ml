## RNN vs LSTM vs GRU

| Architecture | Advantages | Disadvantages |
|-------------|-----------|---------------|
| **RNN** | Simple, fast | Vanishing gradient problem |
| **LSTM** | Remembers long sequences | More parameters, slower |
| **GRU** | Better than RNN, simpler than LSTM | May forget important info |

### LSTM Cell Equations
$$f_t = \sigma(W_f \cdot [h_{t-1}, x_t] + b_f)$$ (forget gate)
$$i_t = \sigma(W_i \cdot [h_{t-1}, x_t] + b_i)$$ (input gate)
$$\tilde{C}_t = \tanh(W_C \cdot [h_{t-1}, x_t] + b_C)$$ (cell candidate)
$$C_t = f_t * C_{t-1} + i_t * \tilde{C}_t$$ (cell state)
$$o_t = \sigma(W_o \cdot [h_{t-1}, x_t] + b_o)$$ (output gate)
$$h_t = o_t * \tanh(C_t)$$ (hidden state)

```python
# LSTM for sequential prediction
class LSTMPredictor(nn.Module):
    """LSTM network for sequence prediction"""
    
    def __init__(self, input_size, hidden_size, num_layers=2, output_size=1):
        super(LSTMPredictor, self).__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, 
                           batch_first=True, dropout=0.2)
        self.fc = nn.Linear(hidden_size, output_size)
    
    def forward(self, x):
        # x shape: (batch, sequence_length, input_size)
        lstm_out, _ = self.lstm(x)
        # Take output from last timestep
        last_out = lstm_out[:, -1, :]
        output = self.fc(last_out)
        return output

lstm = LSTMPredictor(input_size=5, hidden_size=32, num_layers=2)
print("LSTM Architecture:")
print(lstm)
print(f"\nTotal parameters: {sum(p.numel() for p in lstm.parameters())}")
```
