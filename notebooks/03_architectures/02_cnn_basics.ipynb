# Convolutional Neural Networks (CNNs)

Learn how CNNs use convolutions, pooling, and filters to process spatial data.

```python
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt

# Visualize convolution operation
def visualize_convolution():
    # Simple 5x5 input
    x = np.array([[1, 2, 3, 0, 1],
                  [0, 2, 1, 3, 1],
                  [1, 0, 2, 1, 0],
                  [0, 1, 1, 2, 0],
                  [0, 0, 1, 2, 1]], dtype=float)
    
    # 3x3 filter (kernel)
    kernel = np.array([[1, 0, -1],
                      [1, 0, -1],
                      [1, 0, -1]], dtype=float)
    
    print("Input:")
    print(x)
    print("\nKernel (3x3 filter):")
    print(kernel)
    
    # Manual convolution at position (0,0)
    conv_00 = np.sum(x[:3, :3] * kernel)
    print(f"\nConvolution at (0,0): {conv_00}")

visualize_convolution()
```

## CNN Architecture

Basic CNN layers:
1. **Convolution**: Apply filters to extract features
2. **ReLU**: Non-linear activation
3. **Pooling**: Downsample feature maps
4. **Fully Connected**: Classification

```python
class SimpleCNN(nn.Module):
    """Simple CNN for image classification"""
    
    def __init__(self, num_classes=10):
        super(SimpleCNN, self).__init__()
        
        # Convolutional layers
        self.conv1 = nn.Conv2d(3, 32, kernel_size=3, padding=1)
        self.relu1 = nn.ReLU()
        self.pool1 = nn.MaxPool2d(2, 2)
        
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, padding=1)
        self.relu2 = nn.ReLU()
        self.pool2 = nn.MaxPool2d(2, 2)
        
        # Fully connected layers
        self.fc1 = nn.Linear(64 * 8 * 8, 128)
        self.relu3 = nn.ReLU()
        self.fc2 = nn.Linear(128, num_classes)
    
    def forward(self, x):
        # x shape: (batch, 3, 32, 32)
        x = self.conv1(x)
        x = self.relu1(x)
        x = self.pool1(x)  # (batch, 32, 16, 16)
        
        x = self.conv2(x)
        x = self.relu2(x)
        x = self.pool2(x)  # (batch, 64, 8, 8)
        
        x = x.view(x.size(0), -1)  # Flatten
        x = self.fc1(x)
        x = self.relu3(x)
        x = self.fc2(x)
        
        return x

cnn = SimpleCNN(10)
print("CNN Architecture:")
print(cnn)
print(f"\nTotal parameters: {sum(p.numel() for p in cnn.parameters())}")
```
